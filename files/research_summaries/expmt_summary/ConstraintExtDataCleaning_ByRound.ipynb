```python
# Create/Modify binaries for joint dataset

df['mismatch'] = (df['match'] == 0).astype(int)
df['match_'] = (df['match'] == 1).astype(int)
df['control'] = (df['match'] == 2).astype(int)

df['baseline'] = (df['condition']=="Baseline").astype(int)
df['reputation'] = (df['condition']=="Reputation").astype(int)

df['h2'] = (df['round_number'] > 5).astype(int)

df['black'] = (df['race'] == "Black").astype(int)
df['eastasian'] = (df['race'] == "East Asian").astype(int)
df['hispanic'] = (df['race'] == "Hispanic").astype(int)
df['multiracial'] = (df['race'] == "Multi-racial").astype(int)
df['other'] = (df['race'] == "Other").astype(int)
df['southasian'] = (df['race'] == "South Asian").astype(int)
df['white'] = (df['race']=="White").astype(int)

df['asian'] = ((df['eastasian'] == 1) | (df['southasian'] == 1)).astype(int)
df['other_race'] = ((df['multiracial']==1) | (df['other']==1)).astype(int)

df['female'] = (df['gender'] == "Female").astype(int)
df['male'] = (df['gender'] == "Male").astype(int)
df['nonbinary'] = (df['gender'] == "Non-Binary").astype(int)

df['match_h2'] = df['match_'] * df['h2']
df['match_reputation'] = df['match_'] * df['reputation']
df['reputation_h2'] = df['reputation'] * df['h2']
df['match_rep_h2'] = df['match_'] * df['reputation'] * df['h2']
df['trolley_social_action'] = df['trolley'] * df['social_action']

# Save combined data for stata in .csv
# df.to_csv('2020-07-13_spi_strategies_combined_byround_trimmed.csv', encoding='utf-8', index=False)
```

```python
df = df.reset_index(drop=True)

df['switch'] = df.groupby(df['playerid'])['social_action'].apply(lambda x: abs(x - x.shift(1)))
rounds = df['round_number'] == 1 
df.loc[rounds, 'switch'] = df.loc[rounds, 'switch'].fillna(value=0)

df['cum_switch']  = df.groupby(df['playerid'])['switch'].apply(lambda x: x.cumsum())
df['cum_earnings'] = df.groupby(df['playerid'])['round_earnings'].apply(lambda x: x.cumsum())

df.to_csv('PoliticalStances_Combined_ByRound', encoding='utf-8', index=False)
```
