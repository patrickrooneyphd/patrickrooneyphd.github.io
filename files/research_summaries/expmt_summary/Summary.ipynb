## Summary of Political Stances Experiment and Text Analysis

### Why Does Political Stance-Taking Matter as a Business Concern? 

Over the past decade, companies and firm decision makers have begun taking public political stances at a staggering rate. From the conservative owners of [Chic-Fil-A](https://www.cnbc.com/2019/04/02/chick-fil-a-loses-two-airport-deals-in-less-than-two-weeks.html) to the liberal owner of [Penzey's Spices](https://www.newyorker.com/culture/annals-of-gastronomy/ceo-who-called-trump-racist-penzeys-spice-mix), CEOs and other firm decision makers are taking stances that impact their firms' [brand image](https://journals.sagepub.com/doi/full/10.1177/1086026619848144?casa_token=26JiUcwzsmMAAAAA%3AZQ2hRkWuXYKSL6qtODmD7-9dF1WYdxo3S0-n6ttX-fg4zw6vMJu9sFjZ4AsSD2pc8TuSNpIyHIE), [employee motivation](https://pubsonline.informs.org/doi/abs/10.1287/mnsc.2019.3562), and [consumer demand](https://scholar.google.com/citations?user=FQR5W_UAAAAJ&hl=en&oi=sra#d=gs_md_cita-d&u=%2Fcitations%3Fview_op%3Dview_citation%26hl%3Den%26user%3DFQR5W_UAAAAJ%26citation_for_view%3DFQR5W_UAAAAJ%3AUeHWp8X0CEIC%26tzom%3D240).

The business implications of stance-taking are significant. Consider the following three examples:
* In 2017, Pepsi pulled a prominent [advertisement campaign](https://people.com/food/kendall-jenner-pepsi-commercial-company-cost/#:~:text=%E2%80%9CAt%20least%20%242%20million%2C%20but,time%20fee%20for%20the%20ad) costing roughly \$100 Million after it was widely panned for making light of social tensions on racial justice and police reform. 

* In the wake of the 2018 Parkland High School shooting, Ed Stack, CEO of Dick's Sporting Goods, decided to discontinue sales of firearms, which directly led [62 employees to leave the business](https://www.wsj.com/video/dicks-ceo-on-how-taking-a-stand-on-gun-control-hurt-business/CF905F2D-81A8-4B34-9F17-7D59B05350D9.html).

* In the summer of 2020, Facebook confronted a [boycott](https://www.wired.com/story/rashad-robinson-facebook-ad-boycott/) from advertisers who were unhappy with the company's inaction to combat civil rights violations on the site. The total advertisement buys were worth [$7 billion](https://www.wired.com/story/rashad-robinson-facebook-ad-boycott/), representing 10% of Facebook's yearly revenue. In response, the company's COO, Sheryl Sandberg, [took a stance](https://about.fb.com/news/2020/07/civil-rights-audit-report/) against hate speech and instituted a series of multimillion dollar initiatives to police divisive content and promote diversity inside and outside the firm.

While research and real-world examples show that stance-taking has serious implications for the bottom line, less systematic evidence exists on how firm decision makers strategize with respect to stance-taking. Do they always take stances that will yield higher profits for themselves and for the firm? How much do an individual's own beliefs and values matter in this decision? Can we isolate these factors to show their relative causal impacts on decision making? 

This research project aims to understand the factors behind why individual decision makers take political stances in their day-to-day business. By learning more about who takes stances and why, this project aims to better inform corporate decision makers of the dangers and opportunities from taking political stances. It also provides guidance to boards of directors looking to create governance rules around political stance-taking. 

### Experimental Design

In this experiment, I investigate causal factors behind why firm decision makers might take a political stance as part of a business strategy. As a baseline, I focus specifically on: 
1. The match between decision maker beliefs and political causes
2. The monetary payoffs from stance-taking in niche markets
3. External factors that might constrain or enable stance-taking, such as activist pressure. 

As an extension, I also investigate heterogeneous effects by individuals' political and ethical orientations.

The experimental design was inspired by [Ederer and Manso (2013, ManSci)](https://pubsonline.informs.org/doi/abs/10.1287/mnsc.1120.1683?journalCode=mnsc), who examine the impact of incentives on innovation activity. It is a strategy exploration simulation, in which participants serve as the CEO of a hypothetical company and are asked to choose inputs for a product. 

Each bundle of inputs is associated with an unknown profit outcome. Participants explore the landscape of input combinations to try to increase their profits. Below is an example round of the game with associated payoffs.

To guide their exploration, I gave participants two reference strategies and showed the associated profits. The first reference strategy was provided at the beginning of the experiment and showed a strategy that did not involve taking a political stance. After submitting five input bundles and learning their associated profits, participants then observed a second reference strategy (in the form of market research) that involved taking a political stance. This market research hinted at higher monetary benefits from taking a political stance within a certain market niche. 

Participants then submitted five more input bundles and learned associated profits in each of the five remaining periods. Finally, they answered demographic questions, questions about their ethical and political beliefs, and questions about how they made their strategies during the game. 

I ran this study in the summer of 2020 on Prolific, an academic-oriented online platform for subject recruitment. Participants in this study were Americans who self-identified as social conservatives or social liberals. The political causes centered on gun control, which Gallup identified in 2018 as one of the most starkly polarizing issues between Republicans and Democrats. Participants were randomly assigned the opportunity to donate to a social liberal cause ([Everytown for Gun Safety](https://www.everytown.org/)), a social conservative cause ([the National Rifle Association](https://www.nrafoundation.org/)), or a “placebo”, apolitical cause ([the American Red Cross](https://www.redcross.org/)). 

In the study, taking a stance involved donating 10% of earnings to the cause. Earnings were paid out to participants and donations were sent in bulk to the causes after the experiment had ended. Due to these real stakes, participants were incentivized to take both the personal payoffs and costly stances seriously. To allow for testing of external constraints, I ran one additional study in which future profitability fell when participants switched stances, representing credible threats from activists who seek to punish firms that “flip flop” on stances. 

## Findings

Below is an abbreviated summary statistics table that shows feature means, broken down by political orientation (the *liberal* variable). 

Overall, we see that participants take stances in a little over half of the experimental rounds and earn about \$2.50 per round. There is no statistical difference between liberals and conservatives on stance-taking or performance.

There are several hints that participants truthfully answer Prolific's political orientation screening question. First, participants who describe themselves as liberal on Prolific are much more likely to describe themselves as liberal in the experiment, and vice versa. Second, differences in demographic means also align with overall US population differences on this dimension. For example, conservatives in the experiment are older, whiter, and more likely to be male. Finally, conservatives in the experiment are hold conservative beliefs on certain issues; they are more likely to believe in the fairness of the free market and believe that stance taking and social responsibility are less appropriate in business.

```python
from scipy.stats import ttest_ind
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import rpy2.rinterface
import statsmodels.api as sm
from stargazer.stargazer import Stargazer
from IPython.core.display import HTML

df_sumstats = pd.read_csv("PoliticalStances_Baseline_AllRounds.csv")

#stats_list = ['count', 'mean', 'std', 'min', 'max']
stats_list = ['mean']
var_list = ['donate_pct', 'earnings', 'age', 'male', 'white', 'black', 'hispanic', 'asian', 'other_race',
            'volunteer', 'donated', 'politics', 'fair', 'ceostanceopinion', 'ceocsropinion', 'trolley', 'risk', 'amb',
           ]

df_sumdescribe = df_sumstats[var_list]

sumstats = df_sumstats.groupby(df_sumstats['liberal']).agg(
    {"donate_pct": stats_list, 
     "earnings": stats_list,
     "age": stats_list, 
     "male": stats_list, 
     "white": stats_list,
     "black": stats_list, 
     "hispanic": stats_list, 
     "asian": stats_list, 
     "other_race": stats_list, 
     "volunteer": stats_list,
     "donated": stats_list,
     "politics": stats_list, 
     "fair": stats_list,
     "ceostanceopinion": stats_list, 
     "ceocsropinion": stats_list,
     "trolley": stats_list, 
     "risk": stats_list,
     "amb": stats_list
    }
)

# create differences in means between cols as a new col
# df_sumstats['diff_means'] = df_sumstats['mean'][df_sum_stats['liberal']==0] - df_sumstats['mean'][df_sum_stats['liberal']==1]

def append_diff_means(df, varlist, group):
    diff_list = []
    p_val_list = []
    for v in varlist:
        m = df[v][df[group]==1].mean() - df[v][df[group]==0].mean()
        t, p = ttest_ind(df[v][df[group]==1].dropna(), df[v][df[group]==0].dropna(), equal_var=False)
        diff_list.append(m)
        p_val_list.append(p)
    return diff_list, p_val_list

diff_means, p_vals = append_diff_means(df_sumstats, var_list, 'liberal')
```

```python
sumstats_t = sumstats.T
sumstats_t
sumstats_t.rename(index={'donate_pct': "Stance in Round", 
                          'earnings': "Earnings in Round",
                          'age': "Age",
                          'male': "Male",
                          'white': "White",
                          'black': "Black",
                          'hispanic': "Hispanic",
                          'asian': "Asian",
                          'other_race': "Other Race",
                          'volunteer': "Volunteer Prev Yr",
                          'donated': "Donated Prev Yr",
                          'politics': "Social Liberal (1-5)",
                          'fair': "Fair Market (-5 to 5)",
                          'ceostanceopinion': "CEO Stance Approp",
                          'ceocsropinion': "Firm CSR Approp",
                          'trolley': "Outcome Ethics",
                          'risk': "Risk Averse (1 to 5)",
                          'amb': "Ambiguity Averse (1 to 5)",
                         }, inplace=True)
sumstats_t.rename(columns={0: "Conservative", 
                           1: "Liberal",
                         }, inplace=True)
sumstats_t['Diff (Lib - Cons)'] = diff_means
sumstats_t['Diff P-Vals'] = p_vals
sumstats_t
```

This hints at an asymmetric demotivating effect, in which mismatched causes decrease stance taking more than matched causes increase stance taking. Interestingly, it appears that mismatch between the decision maker and the cause can *lower* unprofitable stance-taking when political stance strategies have uncertain outcomes, with the mismatched earning about 20% more over the first five rounds.

### Conclusion and Recommendations

The choice to take a political stance at work involves a complex mix of factors, and the analyses above show that multiple factors causally impact this high stakes decision. Moreover, liberals and conservatives, and those with different ethical outlooks, view stance-taking in different ways.

There can be massive financial implications for taking (or not taking) certain political stances. On a practical level, this research provides guideposts for firms who are looking to manage this critical aspect of their firm's non-market strategy, especially with respect to their executive decision makers. For example, when choosing a new CEO, firms should understand that CEO political and ethical beliefs will likely *causally* impact their approaches to stance-taking. Firms can also use this work to justify governance structures that may constrain or enable stance-taking in the workplace.
