```python
# Further Variable Construction

def cause_function(row):
    if row['cause'] == "Everytown for Gun Safety":
        val = 1
    elif row['cause'] == "The American Red Cross":
        val = 2
    else:
        val = 0
    return val

df['cause_dummy'] = df.apply(cause_function, axis=1)

# Create binaries

df['mismatch'] = (df['match'] == 0).astype(int)
df['match_'] = (df['match'] == 1).astype(int)
df['control'] = (df['match'] == 2).astype(int)

df['nra'] = (df['cause'] == "the National Rifle Association (NRA)").astype(int)
df['etfgs'] = (df['cause'] == 'Everytown for Gun Safety').astype(int)
df['arc'] = (df['cause'] == "The American Red Cross").astype(int)

df['no_stance'] = (df['social_action'] == 0).astype(int)
df['stance'] = (df['social_action'] == 1).astype(int)

df['h2'] = (df['round_number'] > 5).astype(int)

df['sweater'] = (df['type'] == "Sweater Vest").astype(int)
df['cotton_h1'] = (df['cotton'] == 80).astype(int)
df['price_h1'] = (df['price'] == 49).astype(int)

df['cardigan'] = (df['type'] == "Cardigan").astype(int)
df['cotton_h2'] = (df['cotton'] == 40).astype(int)
df['price_h2'] = (df['price'] == 99).astype(int)

df['ref_1'] = (df['sweater'] + df['cotton_h1'] + df['price_h1'] == 3).astype(int)
df['ref_2'] = ((df['cardigan'] + df['cotton_h2'] + df['price_h2'] == 3) & (df['h2'] == 1)).astype(int)

df['ref_1_ct'] = df[['sweater', 'cotton_h1', 'price_h1']].sum(axis=1)
df['ref_2_ct'] = df[['cardigan', 'cotton_h2', 'price_h2']].sum(axis=1)

df['conservative'] = (df['liberal'] == 0).astype(int)

df['rule_based'] = (df['trolley'] == 0).astype(int)
df['conseq'] = (df['trolley'] == 1).astype(int)

df['match_h2'] = df['match_'] * df['h2']
df['mismatch_h2'] = df['mismatch'] * df['h2']
df['control_h2'] = df['control'] * df['h2']

df['trolley_h2'] = df['trolley'] * df['h2']
df['trolley_match'] = df['trolley'] * df['match_']
df['trolley_mismatch'] = df['trolley'] * df['mismatch']

df['black'] = (df['race'] == "Black").astype(int)
df['eastasian'] = (df['race'] == "East Asian").astype(int)
df['hispanic'] = (df['race'] == "Hispanic").astype(int)
df['multiracial'] = (df['race'] == "Multi-racial").astype(int)
df['other'] = (df['race'] == "Other").astype(int)
df['southasian'] = (df['race'] == "South Asian").astype(int)
df['white'] = (df['race']=="White").astype(int)

df['asian'] = ((df['eastasian'] == 1) | (df['southasian'] == 1)).astype(int)
df['other_race'] = ((df['multiracial']==1) | (df['other']==1)).astype(int)

df['female'] = (df['gender'] == "Female").astype(int)
df['male'] = (df['gender'] == "Male").astype(int)
df['nonbinary'] = (df['gender'] == "Non-Binary").astype(int)

df['cum_earnings'] = df.groupby(df['playerid'])['round_earnings'].apply(lambda x: x.cumsum())

save_trimmed_df()
```
